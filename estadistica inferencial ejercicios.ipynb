```python


import pandas as pd
import numpy as np
# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
# Evaluar linealidad de las relaciones entre las variables
# y la distribución de las variables
# ------------------------------------------------------------------------------
#import scipy.stats as stats
import scipy.stats as st
import scipy.stats as stats
from scipy.stats import shapiro, poisson, chisquare, expon, kstest
```

```python
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
# Gestión de los warnings
# -----------------------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("merchant_fill_nonulls.csv")
df.head(2)
```

```python
df_filtered = df[df['rating_count'] >= 500]  #para filtrar por valoraciones
prices = df_filtered['price']
n = len(prices)                    # Tamaño de la muestra
mean_price = prices.mean()         # Media
std_price = prices.std(ddof=1)     # Desviación estándar de la muestra
confidence_interval = stats.t.interval(
    confidence=0.95,           # Nivel de confianza 95%
    df=n-1,                     # Grados de libertad
    loc=mean_price,             # Media de la muestra
    scale=std_price / np.sqrt(n)  # Error estándar de la media
)
print(f"Media del precio: ${mean_price:.2f}")
print(f"Intervalo de confianza al 95%: ${confidence_interval[0]:.2f} - ${confidence_interval[1]:.2f}")
```

* Eres un analista de ventas en una plataforma de comercio electrónico y quieres calcular el intervalo de confianza al 90% para el número promedio de unidades vendidas por producto.

```python
# Columna de unidades vendidas
units = df['units_sold']

# Tamaño de la muestra
n = len(units)

# Media y desviación estándar de la muestra
mean_units = units.mean()
std_units = units.std(ddof=1)  # ddof=1 para desviación muestral

# Calcular el intervalo de confianza al 90%
confidence_interval = stats.t.interval(
    confidence=0.90,         # Nivel de confianza 90%
    df=n-1,                  # Grados de libertad
    loc=mean_units,          # Media de la muestra
    scale=std_units / np.sqrt(n)  # Error estándar de la media
)

# Mostrar el resultado
print(f"Media de unidades vendidas por producto: {mean_units:.2f}")
print(f"Intervalo de confianza al 90%: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

```python

# Crear el histograma
plt.figure(figsize=(10,6))
sns.histplot(units, bins=20, kde=False, color='skyblue')

# Línea de la media
plt.axvline(mean_units, color='red', linestyle='--', linewidth=2, label=f'Media = {mean_units:.2f}')

# Líneas del intervalo de confianza
plt.axvline(confidence_interval[0], color='green', linestyle='--', linewidth=2, label=f'IC 90% = {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='--', linewidth=2, label=f'IC 90% = {confidence_interval[1]:.2f}')

# Etiquetas y título
plt.title('Histograma de unidades vendidas por producto')
plt.xlabel('Unidades vendidas')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
